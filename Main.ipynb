```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df_orig = pd.read_csv('Fortune500.csv')

df=df_orig[:490] #For working on the data
df_new=df_orig[490:] #For predicting values later

X = df[['Rank','Employees','Revenues','Revchange','Prftchange','Assets','Totshequity']].values
y = df[['Profits']].values


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)


reg = LinearRegression()
reg.fit(X_train, y_train)

y_pred=reg.predict(X_test)

r2_score=reg.score(X_train,y_train)
print(r2_score)
```
